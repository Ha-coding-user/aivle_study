# K-Fold Cross Validation Project 1

- K-Fold Cross Validation을 이용해 여러 모델 성능 평가
- Boston Data
- target : medv

```python
# Library 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%config InlineBackend.figure_format='retina'
```

```python
# data 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/boston.csv'
data = pd.read_csv(path)
```

```python
# 상위 5개 행 확인
data.head()
```

```python
# 하위 5개 행 확인
data.tail()
```

```python
# 기초통계량 확인
data.describe()
```

```python
# 상관관계 heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(data.corr(),
            annot=True,
            fmt='.3f',
            vmin=-1, vmax=1,
            square=True,
            cmap='Blues')
plt.show()
```

```python
# target 확인
target = 'medv'

# x, y 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

```python
# train, test 분리
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# KNN을 위한 정규화
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_s = scaler.transform(x_train)
x_test_s = scaler.transform(x_test)
```

```python
# 성능 예측 - Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

model = LinearRegression()
cv_score = cross_val_score(model, x_train, y_train, cv=10)

print(cv_score)
print('평균:', cv_score.mean())
print('표준편차:', cv_score.std())

result = {}
result['LinearRegression'] = cv_score.mean()

result
```

```python
# 성능 예측 - KNN
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score

model = KNeighborsRegressor()
cv_score = cross_val_score(model, x_train_s, y_train, cv=10)

print(cv_score)
print('평균:', cv_score.mean())
print('표준편차:', cv_score.std())

result['KNN'] = cv_score.mean()
result
```

```python
# 성능 예측 - Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

model = DecisionTreeRegressor()
cv_score = cross_val_score(model, x_train, y_train, cv=10)

print(cv_score)
print('평균:', cv_score.mean())
print('표준편차:', cv_score.std())

result['DecisionTree'] = cv_score.mean()
result
```

```python
# 시각화 준비 단계
perf_dic = {'model': result.keys(), 'score': result.values()}
df = pd.DataFrame(perf_dic)
df.sort_values(by='score', inplace=True, ascending=False)
df
```

```python
# 시각화
plt.figure(figsize=(5, 4))
sns.barplot(x=df['score'], y=df['model'])
plt.show()
```

```python
# 성능 평가
from sklearn.metrics import mean_absolute_error, r2_score

model = DecisionTreeRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print('mae:', mean_absolute_error(y_test, y_pred))
print('r2:', r2_score(y_test, y_pred))
```
