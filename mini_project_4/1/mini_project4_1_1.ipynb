**휴일별 데이터 분석**

```python
import requests
import json

def getHolidays(year, month):
    url = 'https://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getRestDeInfo'
    AUTH_KEY = 'VivdmVwNLrv6rXkh5ck2apo6bZrQFYxwj29T4EYvTgVIa53JQgu5qX3UgfK1tw1y1L6cdzfoxkgSMRJr6gPSFA=='
    params = {
        'solYear': str(year),
        'solMonth': str(month).zfill(2),
        '_type': 'json',
        'ServiceKey': AUTH_KEY
    }

    res = requests.get(url, params=params)
    dic = json.loads(res.text)
    counts = dic['response']['body']['totalCount']

    if counts < 1:
        return []
    
    item = dic['response']['body']['items']['item']

    if counts == 1:
        return [item['locdate']]
    
    hol_list = []
    for day in item:
        hol_list.append(day['locdate'])
    return hol_list
```

```python
year_range = range(2017, 2023)
month_range = range(1, 13)

holiday_list = []

for year in year_range:
    for month in month_range:
        holiday_list += getHolidays(year, month)

print(holiday_list)
```

```python
holiday = pd.to_datetime(holiday_list, format='%Y%m%d').tolist()
```

```python
df_total['휴일'] = 0

for i in range(len(df_total)):
    df_total['휴일'][i] = 1 if df_total['기준일ID'][i] in holiday else 0
```

```python
df_holidaySeg = df_total.groupby('휴일').mean()[['총생활인구수']]

axes = sns.barplot(x=df_holidaySeg.index, y='총생활인구수', data=df_holidaySeg)
for p in axes.patches:
    axes.text(p.get_x() + p.get_width()/2,
              p.get_y() + p.get_height(),
              f'{p.get_height():.0f}',
              ha='center')
plt.legend(['0: False', '1: True'], loc='lower right')
plt.show()
```

**Insight**
1. 휴일보다 주말에 총생활인구수가 적은 것을 알 수 있음
